```python
import geopandas as gpd
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import rasterio
import rasterio.plot as rplt
import matplotlib.pyplot as plt
```

```python
data_folder = r"D:\[Code]\HydrofractureShackleton_2023\data"
data_dmg_folder = r"D:\[Code]\HydrofractureShackleton_2023\data-dmg"
data_img_folder = r"D:\[Code]\HydrofractureShackleton_2023\data-img"
figure_folder = r"D:\[Code]\HydrofractureShackleton_2023\figures"
```

```python
lima_files = []
for root, dirs, files in os.walk(data_img_folder):
    for name in files:
        if "RGBREF" in name and name.endswith(".tif"):
            lima_files.append(os.path.join(root, name))
            print(os.path.join(root, name))
```

```python
fig, ax = plt.subplots()
for i_lima, lima_file in enumerate(lima_files):
    if i_lima == 3:
        break
    with rasterio.open(lima_file) as src:
        raster  = src.read()
        transform = src.transform
        print(transform)
        rplt.show(raster, transform = transform, ax=ax)
    
```
