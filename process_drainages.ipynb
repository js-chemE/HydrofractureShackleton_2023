```python
import os
from pprint import pprint
from typing import List

import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

import app
```

```python
filepath = r"D:\[Code]\HydrofractureShackleton_2023\data-hydrofracture\shackleton_hydrofractures_JS.xlsx"
shape_folder = r"D:\[Code]\HydrofractureShackleton_2023\data-hydrofracture"
```

```python
drains = app.load_drain_file(filepath).plot()
```

```python
drains
```

```python
raw = app.combine_shape_files(
    [
        os.path.join(shape_folder, f)
        for f in os.listdir(shape_folder)
        if "L8S2" in f and ".shp" in f
    ]
)
raw["lake id"] = raw["lake id"].astype("int")
raw
```

```python
def get_polygon(row, search, window: int):
    d = search.loc[search["name"] == "tile-" + str(row["tile"])].copy()
    d = d.loc[search["date-0"] == row["date-0"].strftime("%Y-%m-%d")]
    d = d.loc[search["lake id"] == row["lake"]]
    d = d.loc[search["window"] == window]

    if d.shape[0] == 1:
        return d["geometry"].iloc[0]

    return None

window0 = drains.apply(get_polygon, search=raw, window=0, axis=1, result_type="reduce")
window1 = drains.apply(get_polygon, search=raw, window=1, axis=1, result_type="reduce")
```

```python
gpd.GeoSeries(window0[0]).plot()
```

```python
drains_ = drains.copy()
drains_["type"] = "point"

drains0 = drains.copy()
drains0["geometry"] = window0
drains0["type"] = 0
drains0.set_crs(app.ENVIRONMENT.get("Geospatial", "crs_ant_str"), allow_override=True)

drains1 = drains.copy()
drains1["geometry"] = window1
drains1["type"] = 1
drains1.set_crs(app.ENVIRONMENT.get("Geospatial", "crs_ant_str"), allow_override=True)
```

```python
drains_.to_excel(os.path.join(shape_folder, "shackleton_hydrofractures.xlsx"))
drains0.to_excel(os.path.join(shape_folder, "shackleton_hydrofractures_0.xlsx"))
drains1.to_excel(os.path.join(shape_folder, "shackleton_hydrofractures_1.xlsx"))
```
